# Importing Dependences

```python
import mediapipe as mp
import cv2
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

# Function

```python
def create_df_values(result,
                     df,
                     image_file,
                     sign,
                     type_lm,
                     frame):
    #result = results.left_hand_landmarks.landmark
    len_result = len(result)

    source_list = [source for _ in range(len_result)]
    filename_list = [image_file.replace(base_url + "\\", "") for _ in range(len_result)]
    sign_list = [sign for _ in range(len_result)]
    type_list = [type_lm for _ in range(len_result)]
    landmark_index_list = [i for i in range(len_result)]
    x_list = [result[i].x for i in range(len_result)]
    y_list = [result[i].y for i in range(len_result)]
    z_list = [result[i].z for i in range(len_result)]
    frame_list = [frame for i in range(len_result)]

    df = pd.concat([df, pd.DataFrame(data = {"source": source_list,
                                            "filename": filename_list,
                                            "sign": sign_list,
                                            "type": type_list,
                                            "landmark_index": landmark_index_list,
                                            "x": x_list,
                                            "y": y_list,
                                            "z": z_list,
                                            "frame": frame_list})], ignore_index=True)
    
    return df

mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic(static_image_mode=True,
                                model_complexity=2,
                                refine_face_landmarks=True)
```

# Extract Landmarks

```python
base_url = "c:\\Users\\syahr\\Desktop\\Portfolio\\Sign-Language-Recognition\\data"
source = "https://www.kaggle.com/datasets/grassknoted/asl-alphabet"
try:
    df = pd.read_csv("data.csv").iloc[0:0]
except:
    df = pd.DataFrame(data = {"source": [],
                              "filename": [],
                              "sign": [],
                              "type": [],
                              "landmark_index": [],
                              "x": [],
                              "y": [],
                              "z": [],
                              "frame": []})

data_folder = base_url + "\\asl-alphabet\\asl_alphabet_train\\asl_alphabet_train"

for sign in [i for i in os.listdir(data_folder) if i not in ["J", "Z", "space", "del", "nothing"]]:
    sign_folder = data_folder + "\\" + sign

    df_temp = df.copy()
    for image_file in [sign_folder + "\\" + i for i in os.listdir(sign_folder)]:
        image = cv2.imread(image_file)
        results = holistic.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        frame = 1

        if results.left_hand_landmarks:
            df_temp = create_df_values(result = results.left_hand_landmarks.landmark,
                                df = df_temp,
                                image_file = image_file,
                                sign = sign,
                                type_lm = "left_hand",
                                frame = 1)
        
        if results.right_hand_landmarks:
            df_temp = create_df_values(result = results.right_hand_landmarks.landmark,
                                df = df_temp,
                                image_file = image_file,
                                sign = sign,
                                type_lm = "right_hand",
                                frame = 1)

        if results.face_landmarks:
            df_temp = create_df_values(result = results.face_landmarks.landmark,
                                df = df_temp,
                                image_file = image_file,
                                sign = sign,
                                type_lm = "face",
                                frame = 1)

        if results.pose_landmarks:
            df_temp = create_df_values(result = results.pose_landmarks.landmark,
                                df = df_temp,
                                image_file = image_file,
                                sign = sign,
                                type_lm = "pose",
                                frame = 1)
            
    df = df_temp.copy()        
    df = df.loc[~(df["type"].isin(["face"]))]
    df = df.loc[~(df["sign"].isin(["J", "Z"]))]
    df.to_csv("data.csv", index = False, header = True)

df
```
